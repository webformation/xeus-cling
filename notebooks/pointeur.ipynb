```c++
#include <iostream>
using namespace std;
```

On définit une variable i de type entier, initialisée à 3

```c++
int i = 3;
cout << i << endl;
```

L'opérateur & permet de connaitre l'adresse d'une variable.
On peut donc afficher l'adresse de i.

```c++
cout << &i << endl;
```

On définit une variable de type pointeur vers un entier, initialisée avec l'adresse de i

```c++
int *p = &i;
```

Un pointeur est une variable, dont on peut connaitre l'adresse.

```c++
cout << &p << endl;
```

Un pointeur est une variable, dont la valeur est une adresse.

L'affichage de la valeur de p donne l'adresse de i.

```c++
cout << p << endl;
```

Un pointeur peut être déréférencé avec l'opérateur *.

Le déréférencement permet d'accèder à la valeur de la variable pointée.

```c++
cout << *p << endl;
```
